```python
import random
number= random.randint(1,100)
print("welcome to the Guess game!")
print(" i'm thinking about a number between 1 and 100.Can you guess what is it?!")
guessed=False
while not guessed:
    guess=int(input("enter youor guess"))
    if guess < number :
        print("Your guess is too low. Guess again.")
    elif guess > number:
        print("Your guess is too high. Guess again.")
    else:
        print("Congratulations! You guessed the number correctly!"+str(guess))
        guessed = True    
```

```python
numbers = [1, 3, 5]
print('Numbers:', numbers)
even_numbers  = [2, 4, 6]
print('Even numbers:', numbers)
# adding elements of one list to another
numbers.extend(even_numbers)
print('Updated Numbers:', numbers) 
```
