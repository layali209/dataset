```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 1) Load the dataset
file_path = r"C:\Users\layel\Downloads\STEG_BILLING_HISTORY.csv"
df = pd.read_csv(file_path)

# 2) Display the first 10 lines and store them in client_0_bills
client_0_bills = df.head(10)
print("First 10 rows (client_0_bills):")
print(client_0_bills)

# 3) Data type of client_0_bills
print("\nData type of client_0_bills:", type(client_0_bills))

# 4) General information
print("\nGeneral info:")
df.info()

# 5) Number of rows and columns
rows, cols = df.shape
print(f"\nRows: {rows}, Columns: {cols}")

# 6) Number of categorical features (object columns)
categorical_cols = df.select_dtypes(include="object").columns.tolist()
print("\nCategorical columns:", categorical_cols)
print("Number of categorical features:", len(categorical_cols))

# 7) Memory usage (bytes and MB)
memory_bytes = df.memory_usage(deep=True).sum()
memory_mb = memory_bytes / (1024 ** 2)
print(f"\nMemory usage: {memory_bytes} bytes (~{memory_mb:.2f} MB)")

# 8) Missing values inspection
missing_values = df.isnull().sum()
print("\nMissing values per column:")
print(missing_values)

# 9) Handle missing values (numeric -> median, categorical -> mode)
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
for col in numeric_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing values after handling:")
print(df.isnull().sum())

# 10) Descriptive analysis on numeric features
print("\nDescriptive statistics for numeric columns:")
print(df.describe())

# 11) Select bills records for client_id == 'train_Client_0' using 2 methods
# Method 1: boolean indexing
client_0_method1 = df[df["client_id"] == "train_Client_0"]

# Method 2: loc
client_0_method2 = df.loc[df["client_id"] == "train_Client_0"]

print("\nClient train_Client_0 records (method 1) - shape:", client_0_method1.shape)
print("Client train_Client_0 records (method 2) - shape:", client_0_method2.shape)

# 12) Encode 'counter_type' into numeric using LabelEncoder
encoder = LabelEncoder()
df["counter_type_encoded"] = encoder.fit_transform(df["counter_type"].astype(str))
print("\nEncoded 'counter_type' into 'counter_type_encoded'.")

# 13) Delete 'counter_statue' feature
if "counter_statue" in df.columns:
    df.drop(columns=["counter_statue"], inplace=True)
    print("\nDropped column: counter_statue")
else:
    print("\nColumn 'counter_statue' not found in dataset.")

print("\nFinal columns:")
print(df.columns)
```
