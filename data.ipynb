```python
import numpy as np


csv_path = r"C:\Users\layel\Downloads\Loan_prediction_dataset (1).csv"


file = open(csv_path, "r", encoding="utf-8-sig")


data = np.genfromtxt(
    file,
    delimiter=",",
    names=True,        
    dtype=None,        
    encoding=None      
)


file.close()

def _normalize(name: str) -> str:
  
    return "".join(ch for ch in name.lower() if ch.isalnum())


candidates = [
    "loanamount", "loan_amount", "loan amount",
    "totalprice", "total_price", "total price",
    "amount", "loanvalue", "loan_value", "loan value"
]
candidates_norm = [_normalize(c) for c in candidates]


colnames = list(data.dtype.names)  
colnames_norm = [_normalize(c) for c in colnames]

loan_col = None
for want in candidates_norm:
    if want in colnames_norm:
        loan_col = colnames[colnames_norm.index(want)]
        break


if loan_col is None:
    print("Could not auto-detect the loan amount column.")
    print("Available columns are:")
    for c in colnames:
        print(" -", c)
    raise SystemExit("Update the candidates list or set loan_col manually.")

print("Using loan amount column:", loan_col)


loan_amounts = data[loan_col].astype(float)


loan_amounts = loan_amounts[~np.isnan(loan_amounts)]

mean_loan = np.mean(loan_amounts)
median_loan = np.median(loan_amounts)
std_loan = np.std(loan_amounts)

print("\nLoan Amount Statistics")
print("----------------------")
print("Count:", loan_amounts.size)
print("Mean:", mean_loan)
print("Median:", median_loan)
print("Standard Deviation:", std_loan)
```
