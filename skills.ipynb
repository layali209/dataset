Question 2 : 

```python
sample_list = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]

def get_last_element(t):
    return t[-1]

sorted_list = sorted(sample_list, key=get_last_element)
print(sorted_list)  
```

Question 3:

```python
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}

combined = d1.copy()

for key, value in d2.items():
    if key in combined:
        combined[key] += value
    else:
        combined[key] = value

print(combined)  
```

```python
items = [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]

def get_float_value(t):
    return float(t[1])

sorted_items = sorted(items, key=get_float_value, reverse=True)
print(sorted_items)
```
